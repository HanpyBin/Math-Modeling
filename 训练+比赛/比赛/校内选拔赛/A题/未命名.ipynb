```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
```

```python
x_coordinates=[3,1,5,4,0,3,7,9,10,14,2,6,11,15,19,22,21,27,15]
y_coordinates=[2,5,4,7,8,11,9,6,2,0,16,18,17,12,9,5,0,9,19]
z_coordinates=[2.5,2,1.5,1.2,0.85,1.3,1.2,3.3,1.4,1.8,2.5,0.8,1.5,0.9,1.4,2.2,1.8,1.4,1.6]
x_coordinates = np.array(x_coordinates)
y_coordinates = np.array(y_coordinates)
z_coordinates = np.array(z_coordinates)
z_coordinate = z_coordinates/(max(z_coordinates)-min(z_coordinates))
```

```python
plt.grid(linestyle='-.', alpha=0.2)
plt.scatter(x_coordinates,y_coordinates, s=z_coordinates*50, c=z_coordinate, alpha=0.8, cmap=plt.cm.Reds)
plt.scatter(10, 10, marker='*', s = 300, c='r', alpha=0.8)
plt.xlabel('X/km')
plt.ylabel('Y/km')
plt.title('')
plt.savefig("location.png", dpi=300)
```
