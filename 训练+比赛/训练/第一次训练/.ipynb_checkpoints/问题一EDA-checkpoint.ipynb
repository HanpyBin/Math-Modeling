```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("1_obs_model_sta_latlon.csv", encoding="ansi")
data.head()
```

```python
Map = folium.Map(location=[data.lat.values[0], data.lon.values[0]])
site_map = folium.map.FeatureGroup()
for i in range(len(data)):
    site_map.add_child(
        folium.CircleMarker(
            location = [data.lat.values[i], data.lon.values[i]],
            color='yellow',
            fill=True,
            fill_color='red'
        )
    )
Map.add_child(site_map)
Map
```

```python
from folium.plugins import HeatMap
new_map = folium.Map(location=[data.lat.values[0], data.lon.values[0]])
heatdata = data[['lat', 'lon']].values.tolist()
HeatMap(heatdata).add_to(new_map)
new_map
```

```python
data2 = pd.read_csv("2_obs_sta_jja_pre_553_1961_2005.csv", encoding="ansi")
data2.head()
```

```python
def transform_date(x):
    return str(x[0])+'-'+str(x[1])+'-'+str(x[2])
```

```python
data2['date'] = data2[['year', 'mon', 'day']].apply(transform_date, axis=1)
```

```python
data2.head()
```

```python
data2.date = pd.to_datetime(data2.date)
```

```python
data2.head()
```

```python
data2.drop(['year', 'mon','day'], axis=1, inplace=True)
data2.set_index('date', inplace=True)
data2.head()
```
