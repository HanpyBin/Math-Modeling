```python
import numpy as np 
import pandas as pd 
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
#
df = pd.read_csv('PDO.csv')
df.columns=['t','g']
Y0 = df['t']
X0 = df['g']
x=np.array(Y0)
y=np.array(X0)
 
 
 
# statsmodels.api
import statsmodels.api as sm 
lowess=sm.nonparametric.lowess
z=lowess(y,x,frac=0.02, it=0)
plt.plot(x,y)
plt.title('lowess')
plt.plot(z[:,0],z[:,1],color='r',lw=1)
 
# Savitzky-Golay filter  
from scipy.signal import savgol_filter
zs=savgol_filter(y, 9, 3) # window size 51, polynomial order 3
plt.figure()
plt.plot(x,y)
plt.title('Savitzky-Golay filter')
plt.plot(x,zs,color='r',lw=1)
plt.show()
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.gca(projection='3d')
intens = pd.read_excel("data.xlsx", header=None)
X = np.arange(1961,2006,1)
Y = np.arange(0,553,1)
X,Y = np.meshgrid(X,Y)
#Z = intens.T.values
Z = (intens.T.values - intens.min().min())/(intens.max().max()-intens.min().min())
surf = ax.plot_surface(X,Y,Z, cmap="rainbow",rstride = 1, cstride = 1)
#ax.contour(X,Y,Z, zdim='z',offset=-2,cmap='rainbow') #offsetZ
ax.zaxis.set_major_locator(LinearLocator(10))
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel(u"/Y")
ax.set_ylabel(u"")
ax.set_zlabel(u"")
plt.savefig(".png")
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.gca(projection='3d')
rx5data = pd.read_excel("rx5data.xlsx", header=None)
X = np.arange(0,45,1)
Y = np.arange(0,553,1)
X,Y = np.meshgrid(X,Y)
Z = rx5data.T.values
surf = ax.plot_surface(X,Y,Z, cmap="rainbow",rstride = 1, cstride = 1)
ax.contour(X,Y,Z, zdim='z',offset=-2,cmap='rainbow') #offsetZ
ax.zaxis.set_major_locator(LinearLocator(10))
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.savefig("rx5data.png")
```
