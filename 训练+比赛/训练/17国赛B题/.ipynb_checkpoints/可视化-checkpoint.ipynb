```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import pandas as pd
from folium.plugins import HeatMap
sns.set(style="darkgrid")
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
file1 = pd.read_excel("附件一：已结束项目任务数据.xls", names=['index', 'latitude', 'longitude','price','status'])
file1.head()
```

```python
comp = file1[file1.status == 1]
imcomp = file1[file1.status == 0]
```

```python
heat_map = folium.Map(location=[file1.latitude.values[0], file1.longitude.values[1]])
heatdata = file1[['latitude', 'longitude']].values.tolist()
HeatMap(heatdata).add_to(heat_map)
heat_map
```

```python
comp_heat_map = folium.Map(location=[comp.latitude.values[0], comp.longitude.values[1]])
comp_heatdata = comp[['latitude', 'longitude']].values.tolist()
HeatMap(comp_heatdata).add_to(comp_heat_map)
comp_heat_map
```

```python
comp_heat_map.save("completion.html")
```

```python
imcomp_heat_map = folium.Map(location=[imcomp.latitude.values[0], imcomp.longitude.values[1]])
imcomp_heatdata = imcomp[['latitude', 'longitude']].values.tolist()
HeatMap(imcomp_heatdata).add_to(imcomp_heat_map)
#imcomp_heat_map.save("imcompletion.html")
imcomp_heat_map
```

```python
file1.head()
```

```python
sns.countplot(x="price", data=file1, palette="rocket")
plt.xticks(rotation=45)
plt.title("price distribution")
```

```python
f, axes = plt.subplots(2, 1, figsize=(22, 10), sharex=True, sharey=True)
sns.countplot(x="price", data=comp, palette="rocket", ax=axes[0]).set_title("price distribution in completion")
plt.xticks(rotation=45)

sns.countplot(x="price", data=imcomp, palette="rocket", ax=axes[1]).set_title("price distribution in imcompletion")
plt.xticks(rotation=45)
plt.savefig("fig1.png")
```

```python
sns.countplot(x="price", data=file1, hue="status")
plt.xticks(rotation=45)
plt.plot(figsize=(44, 20))
```
