```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import norm
import joypy
import matplotlib.cm as cm
```

```python
result1 = pd.read_excel('result1.xlsx', headers=None, names=['timing'])
result2 = pd.read_excel('result2.xlsx', headers=None, names=['timing'])
result3 = pd.read_excel('result3.xlsx', headers=None, names=['timing'])
```

```python
result1['timing1'] = result2['timing']
result1['timing2'] = result3['timing']
result1.columns = ['1.4','1.5','1.7']
```

```python
fig = plt.figure(figsize=(22, 10))
joypy.joyplot(result1, column=['1.4','1.5','1.7'], colormap=cm.cool, alpha=.5, grid=True)
plt.savefig("分布.png", dpi=300)
```

```python
fig = plt.figure(figsize=(22, 10))
sns.distplot(result1, label='1.4')
sns.distplot(result2, label='1.5')
sns.distplot(result3, label='1.7')
plt.legend()
plt.savefig('频率直方图.png', dpi=300)
```
