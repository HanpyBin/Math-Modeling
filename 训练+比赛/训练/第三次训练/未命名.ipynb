```python
def fill_missing(day, step):
    time_table = pd.date_range(start='2014-06-'+day+' 05:30:00', end='2014-06-'+day+' 22:50:00', freq=step)
    time_table = pd.DataFrame(index=time_table)
    data = pd.read_csv("gps/gps_201406"+day+'.csv', encoding='gbk')
    data = data[['业务时间','经度']]
    data1 = pd.DataFrame([['2014-06-'+day+' 05:30:00',1]], columns=data.columns)
    data = data.append(data1)
    data['业务时间'] = pd.to_datetime(data['业务时间'])
    data.set_index('业务时间', inplace=True)
    data.sort_index(inplace=True)
    data = data[(data.index >= '2014-06-'+day+' 05:30:00') & (data.index <= '2014-06-'+day+' 22:50:00')]
    data.columns = ['cnt']
    
    time_table.reset_index(inplace=True)
    time_table.columns=['业务时间']
    time_table.set_index('业务时间')
    data = pd.merge(time_table, data, on='业务时间',how='outer')
    data.set_index('业务时间', inplace=True)
    data = data.resample(step).count()
    data.interpolate('spline',order=3, inplace=True)
    return data
```

```python
def get_maxF(day, K, g, tstart, t1):
    """
    t1:t'
    tstart:T
    g:gap
    mu:μ
    data:get c(t)
    """
    data = fill_missing(day, t1)
    #print(data)
    data = data[data.index >= tstart]
    t2 = int(t1[:-1])
    mu = 1/t2
    F = [0, g*mu*data.iloc[0].values[0]]
    board = []
    timing = [data.index[0]]
    board.append(g*mu*data.iloc[0].values[0])
    for k in range(2, K):
        timing.append(data.index[k-1])
        board.append(g*mu*data.iloc[k-1].values[0])
        sub = []
        for i in range(1, k):
            sub.append(board[i-1]/(K-i))
        sub = sum(sub)
        F.append(F[-1] + board[-1] - sub)
    timing.append(data.index[K-1])
    F.append(0)
    return F[1:], timing
```

```python
table = pd.date_range(start='2014-06-09 05:30:00', end='2014-06-09 21:30:00', freq='30T')
table = pd.DataFrame(index=table, columns=['Max'])
table.reset_index(inplace=True)
table.columns = ['dating','Max']
table.dating = pd.to_datetime(table.dating)
table.set_index('dating', inplace=True)
```

```python
for ind in table.index[:-1]:
    data, timing = get_maxF('09', 10, 30, ind, '5T')
    df = pd.DataFrame(index = timing)
    df['cnt'] = data
    df = df.resample('1800S').max()
    df.index = pd.to_datetime(df.index)
    for ind1 in df.index:
        if np.isnan(table['Max'][ind1]):
            table['Max'][ind1] = df['cnt'][ind1]
        else:
            table['Max'][ind1] = max(df['cnt'][ind1], table['Max'][ind1])
```

```python
table = pd.date_range(start='2014-06-09 05:30:00', end='2014-06-09 21:30:00', freq='15T')
table = pd.DataFrame(index=table, columns=['Max'])
table.reset_index(inplace=True)
table.columns = ['dating','Max']
table.dating = pd.to_datetime(table.dating)
table.set_index('dating', inplace=True)
for ind in table.index[:-3]: # 改变freq需要修改这里的索引
    data, timing = get_maxF('09', 10, 30, ind, '5T')
    df = pd.DataFrame(index = timing)
    df['cnt'] = data
    df = df.resample('15T').max()
    df.index = pd.to_datetime(df.index)
    for ind1 in df.index:
        if np.isnan(table['Max'][ind1]):
            table['Max'][ind1] = df['cnt'][ind1]
        else:
            table['Max'][ind1] = max(df['cnt'][ind1], table['Max'][ind1])
table.Max = table.Max / 4
plt.plot(table.index, table.Max, label='num')
plt.plot([table.index.min(), table.index.max()], [120, 120], '--', label='threshold')
plt.xticks(pd.date_range(start='2014-06-09 06:00:00', end='2014-06-09 22:00:00', freq='2H'), pd.date_range(start='2014-06-09 06:00:00', end='2014-06-09 22:00:00', freq='2H').hour)
plt.xlabel('时间')
plt.ylabel('人数')
plt.legend()
plt.savefig('rentu.png', dpi=300)
```
